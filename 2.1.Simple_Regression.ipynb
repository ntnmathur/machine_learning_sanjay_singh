# Simple Linear Regression Program

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Find Coefficient Values (b0 and b1)

```python
def find_coef(x,y):
    # Get the count of data points
    n=np.size(x) 
    
    # Get the mean values of vector x and vector y
    mean_x,mean_y=np.mean(x), np.mean(y)
    
    # Calculate cross deviations
    Sxy=np.sum(x*y)-n*mean_x*mean_y
    Sxx=np.sum(x*x)-n*mean_x*mean_x
    
    #Coefficients
    b1=Sxy/Sxx
    b0=mean_y-b1*mean_x
    
    return(b0,b1)
```

### Plot Regression Line

```python
def plt_regression_line(x,y,b):
    plt.scatter(x,y,color="m",marker="o",s=30)
    
    #Predict y
    y_pred=b[0]+b[1]*x
    
    #Plot the regression line
    plt.plot(x,y_pred,color="g")
    
    #Print label
    plt.xlabel("x")
    plt.ylabel("y")
    
    #show plot
    plt.show()
```

### Main Function

```python
def main():
    tips=sns.load_dataset("tips")
    df=pd.DataFrame(tips)
    print(df.head())
    
    #observations
    x=df["total_bill"]
    y=df["tip"]
    
    #Estimate the Coefficient
    b=find_coef(x,y)
    
    #Plot Regression Line
    plt_regression_line(x,y,b)
    
if __name__=="__main__":
    main()
```
