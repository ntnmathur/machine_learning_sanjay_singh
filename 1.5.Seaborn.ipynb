```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Seaborn comes with inbuilt datasets
tips=sns.load_dataset("tips")
df=pd.DataFrame(tips)
df.head()
```

```python
df.shape # rows=244, columns=7
```

Lets try to predict the tip based on the data

```python
# Plot the top amounts distribution
sns.distplot(df["tip"])
```

```python
# Correlation Matrix
corr_matrix=df.corr()
corr_matrix
# This only has the 3 columns as only these 3 columns are numbers
```

```python
# Create the heatmap
sns.heatmap(data=corr_matrix, annot=True)
```

Tip is correlated more with total_bill (0.68) rather than size(0.49)

```python
# Scatter Plot Total Bill Vs Tip
sns.relplot(x="total_bill",y="tip",data=df)
```

```python
# Add another column
sns.relplot(x="total_bill",y="tip",data=df,col="time")
```

```python
sns.relplot(x="total_bill",y="tip",data=df,col="time", hue="smoker")
```

```python
sns.relplot(x="total_bill",y="tip",data=df,col="time", hue="smoker", size="size")
```

```python
sns.lmplot(x="total_bill",y="tip",data=df)
```

```python
sns.lmplot(x="total_bill",y="tip",data=df, col="time")
```

```python
sns.lmplot(x="total_bill",y="tip",data=df, col="time", hue="smoker")
```

Conclusion: Smoker gives higher tips for low amounts and non-smokers give higher tip for high amounts

```python
sns.catplot(x="day",y="tip",data=df,kind="swarm")
# kind tells the volume
```

```python
sns.catplot(x="day",y="tip",data=df,kind="violin")
```
